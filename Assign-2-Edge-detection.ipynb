```python
import cv2
from matplotlib import pyplot as plt
import numpy as np 
from scipy import ndimage 

# https://learnopencv.com/edge-detection-using-opencv/
# Read the original image
img = cv2.imread('377.bmp') 
# Display original image
plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()

# Convert to graycsale
# img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img, (3,3), 0) 
 
# Sobel Edge Detection 
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
# Display Sobel Edge Detection Images
plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel Edge Detection on X axis'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Edge Detection on Y axis'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(sobelxy,cmap = 'gray')
plt.title('Sobel Edge Detection on both X and Y axis'), plt.xticks([]), plt.yticks([])
plt.show()

# Canny Edge Detection
canny = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) # Canny Edge Detection
# Display Canny Edge Detection Image
plt.imshow(canny,cmap = 'gray')
plt.title('Canny Edge Detection'), plt.xticks([]), plt.yticks([])
plt.show()


# Roberts Edge Detection https://www.geeksforgeeks.org/python-opencv-roberts-edge-detection/
img = cv2.imread("377.bmp",0).astype('float64') 
roberts_cross_v = np.array( [[1, 0 ], 
                             [0,-1 ]] ) 
  
roberts_cross_h = np.array( [[ 0, 1 ], 
                             [ -1, 0 ]] ) 
img/=255.0
vertical = ndimage.convolve( img, roberts_cross_v ) 
horizontal = ndimage.convolve( img, roberts_cross_h ) 
roberts = np.sqrt( np.square(horizontal) + np.square(vertical)) 
roberts*=255
# Display Roberts Edge Detection Image
plt.imshow(roberts, cmap = 'gray')
plt.title('Roberts Edge Detection'), plt.xticks([]), plt.yticks([])
plt.show()
```
